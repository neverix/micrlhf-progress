```python
import pandas as pd

df = pd.DataFrame(results)

df_ito = pd.DataFrame(results_ito)
```

```python
tasks = set([x["task"] for x in results])
```

```python
tasks
```

```python
data = [x for x in results if x["task"] == "person_profession"]
```

```python
data = data[-8:]
```

```python
full_data = []

for task in tasks:
    data = [x for x in results if x["task"] == task]
    data = data[-8:]
    full_data.extend([
    {
        "TV": (x["tv_loss"]) / x["zero_loss"],
        "ITO (40)": (x["ito_loss"]) / x["zero_loss"],
        "Recon": (x["recon_loss"]) / x["zero_loss"],
        "Cleaned TV": (x["loss"]) / x["zero_loss"],
        "layer": x["layer"],
    } for l, x in zip(
        [2, 4, 6,8,10,12,14,16],
        data
    )])
```

```python
full_data_ito = []

for task in tasks:
    data = [x for x in results_ito if x["task"] == task]
    data = data[-8:]
    full_data_ito.extend([
    {
        # "TV": (x["tv_loss"]) / x["zero_loss"],
        "ITO (20)": (x["ito_loss"]) / x["zero_loss"],
        # "Recon": (x["recon_loss"]) / x["zero_loss"],
        # "Cleaned TV": (x["loss"]) / x["zero_loss"],
        "layer": x["layer"],
        # "task": task
    } for x in data])
```

```python
df = pd.DataFrame(full_data)
df
```

```python
df_ito = pd.DataFrame(full_data_ito)
```

```python
agg_df = df.groupby(["layer"]).mean().reset_index()
```

```python
agg_df
```

```python
agg_df_ito = df_ito.groupby(["layer"]).mean().reset_index()
```

```python
agg_df["ITO (20)"] = agg_df_ito["ITO (20)"]
```

```python
import plotly.express as px

px.line(agg_df, x="layer", y=["TV", "ITO (40)", "ITO (20)", "Recon", "Cleaned TV"], title="Average relative loss across tasks", labels={"value": "Loss", "variable": "Method", "layer": "Layer"}, width=400, height=500).show()
```
