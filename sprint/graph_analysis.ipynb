```python
import json
```

```python
tasks = ["antonyms", "en_es", "es_en", "present_simple_gerund"]
```

```python
graphs = {}

for task in tasks:
    with open(f"../graph-{task}.json") as f:
        graphs[task] = json.load(f)
```

```python
from collections import defaultdict
nodes = defaultdict(dict)

for task in tasks:
    for node in graphs[task]["nodes"]:
        node_id = tuple(node[:4])
        nodes[task][node_id] = node[-1]
```

```python


node_ies = defaultdict(list)

for task in nodes:
    for n_id, w in nodes[task].items():
        node_ies[n_id].append(w)
        
```

```python
[(k, x) for k, x in node_ies.items() if len(x) == len(tasks) and min([abs(y) for y in x]) > 0.005]
```

```python
mean_node_ies = {
    k: sum([abs(x) for x in node_ies[k]]) / 3 for k in node_ies
}
```

```python
mean_node_ies
```

```python
normalized_node_ies = {}

for task in nodes:
    normalized_node_ies[task] = {}
    for node in nodes[task]:
        node_id = tuple(node[:4])
        normalized_node_ies[task][node_id] = mean_node_ies[node_id] / nodes[task][node_id]
```

```python
all_normalized_node_ies = []

for task in normalized_node_ies:
    all_normalized_node_ies.extend([
        (task, k, v) for k, v in normalized_node_ies[task].items()
    ])
```

```python
all_normalized_node_ies = sorted(all_normalized_node_ies, key=lambda x: x[-1], reverse=True)
```
